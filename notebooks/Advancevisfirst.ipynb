```python
# reading the data
def load_data(url): 

    return pd.read_csv(url)
```

```python
def extract(df):

    data = df.drop(df[(df['x_coordinate'] < -89) | (df['x_coordinate'] > 89) | (df['y_coordinate'] < -42.5) | (df['y_coordinate'] > 42.5)].index)
    data['x_coordinate'] = abs(abs(data['x_coordinate']) - 89)
    number_of_games = data['gameID'].nunique()
    
    data = data.groupby(['x_coordinate','y_coordinate']).count()
    
    data = pd.DataFrame(data.iloc[:,0])
    data = data.rename(columns={'event_type':'shot_rate'})

    data['shot_rate'] = data['shot_rate'] / number_of_games

    
    return data
```

```python
def excess_shot_rate(df, year, team):
    data = df

    data_season = data[data['season'] == year * 10000 + year + 1]
    data_team = data_season[data_season['team'] == team]


    league_shot_rate = extract(data_season)/2
    team_shot_rate = exteract(data_team)
    
    return team_shot_rate.sub(league_shot_rate, fill_value=0)
```
